```python
df_dropped.fillna(df_dropped.median(), inplace=True)
df_dropped
```

```python
df_corr = pd.read_csv('Dropped_test_final.csv')
df_corr
```

```python
df_corr = df_corr.drop(["Unnamed: 0", "continent", "total_cases", "new_cases", "total_deaths", "total_cases_per_million", "total_deaths_per_million", "hospital_beds_per_thousand" ], 1)

df_corr
```

```python
col_names = []
obesity = []
undernourished = []
total_deaths = []
deaths = []
death_rate = []

i = 0
while i < 32 :

  
  r_o, p_o = pearsonr(df_corr.iloc[:, i], df_corr.obesity)
  r_u, p_u = pearsonr(df_corr.iloc[:, i], df_corr.undernourished)
  # r_td, p_td = pearsonr(df_dropped.iloc[:, i], df_dropped.total_deaths)
  r_d, p_d = pearsonr(df_corr.iloc[:, i], df_corr.deaths)
  r_dr, p_dr = pearsonr(df_corr.iloc[:, i], df_corr.Death_Rate)

  colname = df_corr.columns[i]

  col_names.append(colname)
  obesity.append(round(r_o, 5))
  undernourished.append(round(r_u, 5))
  # total_deaths.append(round(r_td, 5))
  deaths.append(round(r_d, 5))
  death_rate.append(round(r_dr, 5))

  i+= 1

df_corr_new = pd.DataFrame({'Factors': col_names, 'Obesity': obesity, 'Undernourished': undernourished, 'Deaths': deaths, 'Death Rate': death_rate})

df_corr_new

```

```python
df_corr_new.to_csv("Heatmap_Data_revised.csv")
```
